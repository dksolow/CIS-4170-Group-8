```python
# Importing Packages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# YASHNA

# Inflation Rate Dataframe
inflation_rate_df = pd.read_excel('Inflation Data.xlsx')
inflation_rate_df = inflation_rate_df.set_index('Year')

inflation_rate_df
```

```python
# Inflation Rate Line Plot
plt.figure()
sns.lineplot(x = inflation_rate_df.index, y = inflation_rate_df['Avg'], color = 'b')
plt.title('Inflation in the U.S. from 2000-2021')
plt.xlabel('Year')
plt.ylabel('Average Inflation Rate')
plt.show()
```

```python
# Tuition Dataframe
tuition_df = pd.read_excel('Tuition Data.xlsx')
tuition_df = tuition_df.set_index('Year')

tuition_df
```

```python
# Tuition Line Plot
plt.figure
sns.lineplot(x = tuition_df.index, y = tuition_df['All Institutions'], color = 'r')
plt.title('Average Tuition in the U.S. from 2000-2020')
plt.xlabel('Year')
plt.ylabel('Average Dollars Spent on Tuition')
plt.xticks(rotation = 45)
plt.show()
```

```python
# DAKOTA

# Public vs. Private Institution Tuition Dataframe
private_public_df = pd.read_excel('Public vs Private Constant Dollars.xlsx')
private_public_df
```

```python
# Public vs. Private Institution Tuition Line Plot
private_public_df = private_public_df.loc[33:52]

fig, ax = plt.subplots()

PI_4_year = ax.plot(private_public_df['Year'], private_public_df['PI: 4-year'])
PR_4_year = ax.plot(private_public_df['Year'], private_public_df['PR: 4-year'])

ax.legend(['Public', 'Private'])

plt.xlabel('Year', fontsize = 13)
plt.ylabel('Cost of Tuition', fontsize = 13)
plt.title('Public vs Private 4-year Tuition', fontsize = 15)

plt.xticks(rotation = 45)

ax.set_ylim(ymin = 0)

plt.show()
```

```python
# CPI Inflation Dataframe
cpi_inflation_df = pd.read_excel('Inflation Rate.xlsx')
cpi_inflation_df
```

```python
# CPI Inflation Bar Plot
ax = sns.barplot(x = "Avg-Avg", y = "Year", data = cpi_inflation_df, orient = 'h')
ax.set_xlabel('Average', fontsize = 13)
ax.set_ylabel('Year', fontsize = 13)
ax.set_title('Average CPI for each Year', fontsize = 15)
plt.show()
```

```python
# COVID-19 Inflation Dataframe
covid19_cpi_df = pd.read_excel('Testing Inflation Months.xlsx')
covid19_cpi_df
```

```python
# COVID-19 Inflation Horizontal Bar Plot
x = np.arange(len(covid19_cpi_df['Month']))

fig, ax = plt.subplots(figsize = (15, 8))

inf_2018 = ax.barh(x - 0.4, covid19_cpi_df[2018], 0.2, align = 'center')
inf_2019 = ax.barh(x - 0.2, covid19_cpi_df[2019], 0.2, align = 'center')
inf_2020 = ax.barh(x + 0, covid19_cpi_df[2020], 0.2, align = 'center')
inf_2021 = ax.barh(x + 0.2, covid19_cpi_df[2021], 0.2, align = 'center')

plt.yticks(x, covid19_cpi_df['Month'])
plt.legend(['2018', '2019', '2020', '2021'])
plt.title('COVID-19 Inflation')
plt.xlabel('Inflation', fontsize = 20)
plt.ylabel('Months', fontsize = 20)

plt.gca().invert_yaxis()

plt.show()
```
