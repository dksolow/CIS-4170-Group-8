```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('pu_vs_pr_constant_dollars.xlsx')
```

```python
df
```

```python
#PI_4_year = df.iloc[df[33:52], df[1]]
df_2000 = df.loc[33:52]
```

```python
fig, ax = plt.subplots()


PI_4_year = ax.plot(df_2000['Year'], df_2000['PI: 4-year'])
PR_4_year = ax.plot(df_2000['Year'], df_2000['PR: 4-year'])

ax.legend(['Public', 'Private'])

plt.xlabel('Year', fontsize=13)
plt.ylabel('Cost of Tuition', fontsize=13)
plt.title('Public vs Private 4-year Tuition', fontsize=15)

#plt.xticks(np.arange(0, 53, step=10), rotation = 45)
plt.xticks(rotation = 45)
#plt.yticks(np.arange(0, 40000, step=1))

ax.set_ylim(ymin=0)

plt.savefig("Public vs Private 4-year", dpi = 120, bbox_inches='tight')

plt.show()
```

```python
Average_tuition = pd.read_excel('tuition_change.xlsx')
```

```python
Average_tuition
```

```python
year_len = np.arange(len(Average_tuition['Year']))
```

```python
fig, ax = plt.subplots()

ax.bar(year_len - 0.2, Average_tuition['One Year Change: Private'], 0.2, align='center')
ax.bar(year_len + 0, Average_tuition['One Year Change: Public'], 0.2, align='center')

plt.xticks(year_len, Average_tuition['Year'])
plt.legend(['Private One Year Change', 'Public One Year Change'])
plt.title('Average Tuition Yearly Change')
plt.xlabel('Year')
plt.ylabel('Percentage Change')
plt.xticks(rotation=30)

plt.style.use('ggplot')

plt.show()
```

```python
New_york_tuition = pd.read_excel('tuition_change.xlsx', sheet_name=['New York public vs private'])

New_york_tuition

New_york_tuition = New_york_tuition.get('New York public vs private')

New_york_tuition
```

```python
NY_no_percent_change = New_york_tuition.drop([18, 19])
```

```python
NY_year_len = np.arange(len(NY_no_percent_change['Year']))
```

```python
fig, ax = plt.subplots()

ax.plot(NY_no_percent_change['Year'], NY_no_percent_change['Public Two-Year In-District Tuition and Fees'])
ax.plot(NY_no_percent_change['Year'], NY_no_percent_change['Public Four-Year In-State'])

plt.legend(['Public Two-Year', 'Public Four-Year'])
plt.xticks(rotation=45)
plt.title('New York Public Two vs Four Year')
plt.xlabel('Year')
plt.ylabel('Average Tuition')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(NY_year_len - 0.2, NY_no_percent_change['Public Two-Year In-District Tuition and Fees'], 0.4, align='center')
ax.bar(NY_year_len + 0.2, NY_no_percent_change['Public Four-Year In-State'], 0.4, align='center')

plt.legend(['Public Two-Year', 'Public Four-Year'])
plt.xticks(NY_year_len, NY_no_percent_change['Year'], rotation=45)
plt.title('New York Public Two vs Four Year')
plt.xlabel('Year')
plt.ylabel('Average Tuition')


plt.show()
```
